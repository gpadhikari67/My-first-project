# Gravitation

## Newton's Law of gravitation

### Statement: 
The gravitational force of attraction betweem two mass is directly propertional to the product of its mass and inversely proportional to the squre of the distance from their centre.
#### Let $ m_1 $ and $ m_2 $ be the two mass seperated by distance $r$ from their centre.
   $F$ = $G\frac{m_1m_2}{r^2}$
   
   

```python
import numpy as np
import matplotlib.pyplot as plt
r= np.arange(1,100,1) # in meter
m_1=100 # in kg
m_2=100 # in kg
G=6.67*10**-11 # Nm^2/kg^2
F = G*m_1*m_2/r**2
plt.plot(r,F)
plt.xlabel('Distance (in m)',color='g',fontsize=20)
plt.ylabel('Gravitational Force (in N)',color='r',fontsize=20)
```
