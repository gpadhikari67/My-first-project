# Gravitation
## Newton's Law of Gravitation
### Statement
The gravitational force of attraction between two objects having  masses $m_1 $ and $m_2$ seperated by distance $ r $ from their center is 
 
## $ F = \frac {G m_1m_2}{r^2}   $
where,G is the constant called Universal Gravitational constant having value $6.67 \times 10^{-11}Nm^2/kg^2$ m and M are the masses of two objects and r is the distance between centers of two masses.



![image](https://d1whtlypfis84e.cloudfront.net/guides/wp-content/uploads/2019/09/29192122/Gravitational-force-768x538.png)

```python
import numpy as np
import matplotlib.pyplot as plt
r= np.arange(1,100,1) # in meter
m_1=100 # in kg
m_2=100 # in kg
G=6.67*10**-11 # Nm^2/kg^2
F = G*m_1*m_2/r**2
plt.plot(r,F)
plt.xlabel('Distance (in m)',color='g',fontsize=20)
plt.ylabel('Gravitational Force (in N)',color='r',fontsize=20)
```
